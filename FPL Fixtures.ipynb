```python
import pandas as pd
import numpy as np
import requests 
```

#### In order to get all the fixtures, starting from week 1, you have to add the extension 'future =0'

```python
url = 'https://fantasy.premierleague.com/api/fixtures?future=0'
```

```python
r = requests.get(url)
```

```python
json = r.json()
```

```python
def dict(list):
    for i in json:
        return i.keys()
```

```python
json1 = dict(json)
```

```python
print(json1)
```

```python
fixtures = pd.DataFrame(json)
```

```python
fixtures.head(10)
```

##### You can then covert your dataframe to a CSV file. You can head over to excel to map the club names to their numbers. So for the fixtures schedule you only need the team_a (Away Team) and team_h columns (Home Team).

```python
fixtures.to_csv('Fixtures.csv')
```
